```python
import os
import pandas as pd
import json

cwd = 'C:\\Users\\Galo\\Documents\\Matematicas\\4 Mates\\TFG\\Results fase 1 definitivos\\'
datasets_names = ["mnist_with_clothes","mnist_numbers","chinese_mnist","gamma_telescope","wine","image_segmentation","digits_dataset","breast_cancer"]
ml_algorithms = ["SGDClassifier","gaussianNB","random_forest","mlp_classifier","logistic_regression"]
file_names = os.listdir(cwd)
datasets = []
algorithms = []
ruido = []
sin_ruido = []
con_ruido = []
PvsP = []
PvsD = []
DvsP = []
DvsD = []

for file_name in os.listdir(cwd):
    with open(cwd + file_name, 'r') as file:
        data = json.load(file)
    datasets.append(data["dataset"])
    algorithms.append(data["metodo"])
    ruido.append(data["parametros"][0])
    sin_ruido.append(data["sin_ruido"]["kappas"])
    con_ruido.append(data["con_ruido"]["kappas"])
    PvsP.append(data["PvsP"]["kappas"])
    PvsD.append(data["PvsD"]["kappas"])
    DvsP.append(data["DvsP"]["kappas"])
    DvsD.append(data["DvsD"]["kappas"])
               
datasets_ml = pd.DataFrame({"dataset": datasets, "ml_algorithms": algorithms, "ruido":ruido, \
                                               "sin_ruido":sin_ruido,"con_ruido":DvsP,  \
                                             "PvsP":PvsP,"PvsD":PvsD,"DvsD":con_ruido})
datasets_ml = datasets_ml.iloc[:,[2,1,0,4,5,3,6,7]]
datasets_ml = datasets_ml.rename(columns={"ruido": "Ruido","ml_algorithms":"Algoritmo","dataset":"Dataset","con_ruido":"Con Ruido","sin_ruido":"Sin Ruido"})
datasets_ml
```

```python
import numpy as np
import pandas as pd
from scipy.stats import bootstrap

noise = [0.15,0.3,0.45]
ml_algorithms = ["SGDClassifier","gaussianNB","random_forest","mlp_classifier","logistic_regression"]
xlabel = ["Con Ruido","Con Ruido","PvsP"]
ylabel = ["Sin Ruido","PvsP","Sin Ruido"]
labels = ["Datasets","Con Ruido/Sin Ruido","Con Ruido/PvsP","PvsP/Sin Ruido"]
datasets_names = ["mnist_with_clothes","mnist_numbers","chinese_mnist","gamma_telescope","wine","image_segmentation","digits_dataset","breast_cancer"]
df_mean_IC = {}

def mean_diff(data1, data2):
    return (data1.mean() - data2.mean())
cont = 1
for i in range(5):
    for j in range(3):
        # print(cont)
        cont = cont +1
        dic = {key:[] for key in labels}
        # df_mean = pd.DataFrame(dic)
        df_IC = pd.DataFrame(dic)
        # print(df_mean)
        # print(df_IC)
        # df_IC[labels[0]] = datasets_names

        for k in range(8):
            IC = []
            IC.append(datasets_names[k])
            for pos in range(1,4):
                # media = []
    
                x = datasets_ml.loc[(datasets_ml["Ruido"] == noise[j]) & (datasets_ml["Algoritmo"] == ml_algorithms[i]) & (datasets_ml["Dataset"] == datasets_names[k]),][xlabel[pos-1]]
                y = datasets_ml.loc[(datasets_ml["Ruido"] == noise[j]) & (datasets_ml["Algoritmo"] == ml_algorithms[i]) & (datasets_ml["Dataset"] == datasets_names[k]),][ylabel[pos-1]]
                    
                # print(x.to_numpy()[0])
                # print(x_rand.iloc[i])
                # print(y_rand.iloc[i])
                boot_result = bootstrap((np.array(y.to_numpy()[0]), np.array(x.to_numpy()[0])), mean_diff,  confidence_level=0.95)
                # media.append(round(mean_diff(y, x),4))
                IC.append([round(boot_result.confidence_interval.low,4),round(boot_result.confidence_interval.high,4)])
            dic2 = {labels[pos2] : [IC[pos2]] for pos2 in range(len(labels))}
            df_IC2 = pd.DataFrame(dic2)
            df_IC = pd.concat([df_IC, df_IC2], ignore_index=True)
        print(f"{noise[j]}"+"_"+ml_algorithms[i])
        print(df_IC)
        df_mean_IC[f"{noise[j]}"+"_"+ml_algorithms[i]] = {"IC":df_IC}
        # df_mean_IC[f"{noise[j]}"+"_"+ml_algorithms[i]] = {"mean":df_mean,"IC":df_IC}

df_mean_IC
```
